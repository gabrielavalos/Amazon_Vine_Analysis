```python
# IMPORT DEPENDENCIES
import pandas as pd
```

```python
# READ IN POSTGRES VINE_TABLE CSV AS DATAFRAME df
df = pd.read_csv("vine_table.csv")
```

```python
#df
```

```python
# CREATE NEW DF WHERE total_votes are >= 20
df_total_votes_20 = df.loc[df["total_votes"] >= 20]
```

```python
#df_total_votes_20
```

```python
# CREATE A NEW DF WHERE helpful_votes DIVIDED BY total_votes IS EQUAL >= 50%
df_helpful_by_total = df_total_votes_20.loc[df_total_votes_20['helpful_votes'] / df_total_votes_20['total_votes'] * 100 >= .5]
```

```python
#df_helpful_by_total
```

```python
# CREATE A DF FOR VINE PROGRAM MEMBERS AND NON VINE MEMBERS MEMBERS, RESPECTIVELY
df_yes_vine = df_helpful_by_total.loc[df_helpful_by_total["vine"] == 'Y']
df_no_vine = df_helpful_by_total.loc[df_helpful_by_total["vine"] == 'N']
```

```python
#df_yes_vine
```

```python
#df_no_vine
```

```python
# Determine the total number of reviews, the number of 5-star reviews, 
#and the percentage of 5-star reviews for the two types of review (paid vs unpaid).
paid_vine_total_reviews = df_yes_vine['star_rating'].count()
paid_vine_5star_reviews = df_yes_vine[df_yes_vine['star_rating'] == 5.0]['star_rating'].count()
paid_vine_5star_percentage = (paid_vine_5star_reviews / paid_vine_total_reviews * 100)
```

```python
unpaid_vine_total_reviews = df_no_vine['star_rating'].count()
unpaid_vine_5star_reviews = df_no_vine[df_no_vine['star_rating'] == 5.0]['star_rating'].count()
unpaid_vine_5star_percentage = (unpaid_vine_5star_reviews / unpaid_vine_total_reviews * 100)
```
